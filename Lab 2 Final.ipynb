$\textbf{Lab 2: Vapor Pressure of Unknown Liquid as a Function of Temperature}$

$\textbf{Abstract}$

The purpose of this lab was to utilize a digital pressure gage to take the changes in pressure of cyclohexane, ethanol, and water respectively as a function of temperature, and to use this to find $\Delta H_{vaporization}$, the enthalpy of vaporization, of each liquid. By taking the pressures and temperatures over multiple steps, we can calculate a change in enthalpy via the Classius-Clapeyron equation, which, after creating a Pressure vs. Temperature graph, we can attribute to each plot point. The key findings from this experiment overall showed a decline in enthalpy preportional to the decreases in temperature and vapor pressure, but when compared to theoretical vapor pressures, displayed troubling results.



$\textbf{Introduction}$

The state function, enthalpy, is described as being the sum of the internal energy and pressure multiplied by volume, and is a useful in determining the direction that heat flows in a reaction.$^{1}$  In this lab, we are calculating the entropy of vaporization, which essentially is a value related to the amount of energy necessary to convert liquid particles into the gas phase. To calculate this value, we utilize the Clausius-Clapeyron equation.

The Clausius-Clapeyron equation is a useful equation in relating changes in vapor pressure to changes in temperature given the enthapy of vaporization. In the case of this experiment, the changes in pressure and the changes in temperature are already known, but this equation can still be utilized to find the unknown enthalpy of vaporization for each liquid based on such changes. The relation depicted by this equation is as follows:

$$\ln\Big(\frac{P_{f}}{P_{i}}\Big) = \frac{\Delta H_{vaporization}}{R} \Big(\frac{1}{T_{f}} - \frac{1}{T_{i}}\Big)$$

In this experiment, the enthalpies were calculated by decreasing the temperature and allowing the pressure to stabilize. This allowed for the DPG to measure a more accurate vapor pressure for each respective temperature. This, in turn, would allow us to calculate a more accurate enthalpy of vaporization. 

To carry out the actual experiment, three separate liquids (ethanol, cyclohexane, and water) were studied by decreasing the temperature and observing the change in vapor pressure via the DPG. Because vapor pressure decreases with temperature$^{2}$, and we know that these values should be proportional to changes through $
P \propto e^{\frac{\Delta H_{vaporization}}{RT}}$, we know that the enthalpy should also decrease.$^{3}$ After completing this lab, and calculating the enthalpies of vaporization for each liquid, we found that this did in fact occur as predicted.

```python
# Literature vapor pressures vs temperatures for water, ethanol, cyclohexane

watert = [370.0, 360.0, 353.0, 348.0, 343.0, 330.0, 323.0, 318.0]
watervp = [682.1, 468.7, 355.1, 289.1, 233.7, 129.8, 92.5, 71.9]
# (4) http://intro.chem.okstate.edu/1515SP01/Database/VPWater.html
ethanolt = [349.3, 327.54, 319.67, 301.7, 297.0, 292.8, 290.6]
ethanolvp = [701.1, 273.15, 187.3, 72.6, 55.4, 43.18, 37.75]
# (5) http://ddbonline.ddbst.com/AntoineCalculation/AntoineCalculationCGI.exe?component=Ethanol
cyclot = [348.0, 339.85, 332.6, 321.7, 302.4, 293.4, 291.0, 289.5, 288.0]
cyclovp = [637.9, 490.9, 384, 258.96, 118.6, 79, 70.5, 65.6, 61]
# (6) http://ddbonline.ddbst.com/AntoineCalculation/AntoineCalculationCGI.exe

# Group 1: Ethanol
g1t = [349.3, 327.54, 319.67, 301.7, 297.0, 292.8, 290.6]
g1p = [750, 407, 349, 244, 241, 242, 246]
# Group 2: Cyclohexane
g2t = [348.0, 339.85, 332.6, 321.7, 302.4, 293.4, 291.0, 289.5, 288.0]
g2p = [754, 635, 509, 393, 315, 297, 296, 295, 294]
# Group 3: Water
g3t = [370.0, 360.0, 353.0, 348.0, 343.0, 330.0, 323.0, 318.0]
g3p = [720, 637, 575, 533, 434, 392, 327, 303]
r = 8.314

## ln(Pvapor) = deltaH/R*(1/Tvapor) + C

def enth(t, p):
    #ln(pf/pi) = deltaH/R*(1/Tf - 1/Ti)
    h = -1*np.log(p)*r/t
    return h
    
def PandT(t, p):
    pnew = [y/x for x,y in zip(p, p[1:])]
    tnew = [(1/y) - (1/x) for x,y in zip(t, t[1:])]
    return enth(tnew, pnew)
    

g1H = PandT(g1t,g1p)
print("Group 1 Entropies: ",g1H)
g2H = PandT(g2t, g2p)
print("Group 2 Entropies: ",g2H)
g3H = PandT(g3t, g3p)
print("Group 3 Entropies: ",g3H)

waterH = PandT(watert, watervp)
print("\nTheoretical Water Entropy: ", waterH)
ethanolH = PandT(ethanolt, ethanolvp)
print("Theoretical Ethanol Entropy: ", ethanolH)
cycloH = PandT(cyclot, cyclovp)
print("Theoretical Cyclohexane Entropy: ", cycloH)
```

```python
plt.plot(g1t, g1p, '-b*', )
plt.xlabel('Temperature(K)')
plt.ylabel('Pressure(mmHg)')
plt.title('Group 1 Temperature vs. Pressure\n Ethanol')
plt.show()
```

```python
plt.plot(g2t, g2p, '-b*', )
plt.xlabel('Temperature(K)')
plt.ylabel('Pressure(mmHg)')
plt.title('Group 2 Temperature vs. Pressure\n Cyclohexane')
plt.show()
```

```python
plt.plot(g3t, g3p, '-b*', )
plt.xlabel('Temperature(K)')
plt.ylabel('Pressure(mmHg)')
plt.title('Group 3 Temperature vs. Pressure\n Water')
plt.show()
```

```python
# Errors between experimental and literature vapor pressures
ethanolError = [round((x-y)/y*100, 2) for x,y in zip(g1p, ethanolvp)]
print("Ethanol % Difference: ", ethanolError)
waterError = [round((x-y)/y*100, 2) for x,y in zip(g3p, watervp)]
print("Water % Difference: ", waterError)
cycloError = [round((x-y)/y*100, 2) for x,y in zip(g2p, cyclovp)]
print("Cyclohexane % Difference: ", cycloError)

etdif = [round(abs(x-y),2) for x,y in zip(ethanolH, g1H)]
print("\nEthanol Enthalpy Difference: ", etdif)
waterdif = [round(abs(x-y),2) for x,y in zip(waterH, g3H)]
print("Water Enthalpy Difference: ", waterdif)
cyclodif = [round(abs(x-y),2) for x,y in zip(cycloH, g2H)]
print("Cyclohexane Enthalpy Difference: ", cyclodif)
```

As can be seen from the cell above, the percent error is rather small initially, but becomes very large as the temperature drops in all three liquids, reaching as high as 551%. This raises concerns regarding the validity of the results gathered from this experiment, as our measured vapor pressures should have been significantly lower if going by the theoretical values for each liquid. To illustrate exactly how great of a difference the changes in vapor pressure have on the enthalpy, the difference between the theoretical and experimental enthalpies of vaporization were calculated above, with the greatest difference totaling to over 48.5kJ. 

Besides the differences in experimental and theoretical vapor pressures, there were other sources of error present within this experiment.
