$\textbf{Lab 3: Enthalpy of Mixing}$

$\textbf{Abstract}$

The purpose of this experiment was to observe the effects that mixing a reaction, resulting in either a net breakage of a net formation of H-bonds between reactants occurs, has on the overall Gibbs Free Energy of the reaction through association with enthalpy. This experiment was performed by first determining the heat capacity of the calorimeter using only water to determine it, followed by three subsequent reactions utilizing mixtures of water-ethanol, cyclohexane-acetone, and cyclohexane-ethanol to determine the free energy, the enthalpy, and the entropy of each mixture to find the thermodynamic favorability of each reaction. For the first reaction, the equation $q_{p} = C_{p}\Delta T$ , we can relate the heat lost by the water to the heat gained by the calorimeter, which allowed us to calculate the heat capacity of the calorimeter for usage in the second experiment. In the second experiment, we related the densities and volumes of each reactant to find their masses, and utilized their specific heats, masses, and changes in temperature to find a total heat that could tell us the overall net $\Delta H_{mix}$ for the overall reaction.

$\textbf{Introduction}$

The spontaneity of a reaction can be traced back to the equation for Gibbs Free Energy, $\Delta G = \Delta H + T \Delta S$ when at constant temperature and pressure.$^{1}$ It is through this equation that one can predict the spontaneity of a reaction through noting the direction that $\Delta G$ changes. A positive change in $\Delta G$ predicts that the reaction is nonspontaneous, a negative change indicates spontaneity, and a value of 0 indicates equilibrium of the system. In an ideal solution, $\Delta H_{mix} = 0$ , which ultimately allows us to reform the equation into 
$\Delta G_{mix} = RT\Big( x_{1}\ln x_{1} + x_{2}\ln x_{2}\Big)$ .$^{2,3}$  Using this equation, we will find an free energy for an ideal mixture, which we can compare to the values received experimentally. The experimental total heat can be calculated utilizing the equation: $- q_{total} = m_{Hot} C_{S} \Big( T_{Warm} - T_{Hot} \Big)$
By subtracting our ideal values form our experimental values for total heat, we will yield a value for $\Delta H_{mix}$

As far as the actual methodology of this experiment goes, the experiment was broken down into a two day procedure, with the first portion being geared towards studying the calorimeter, and the latter portion being focused on the actual collection of enthalpy values across three sub-reactions mentioned in the abstract. The former experiment consisted of essentially mixing together two portions of water at differing temperatures spinning inside of a calorimeter and measuring the changes in temperature. Knowing that the calorimeter will ultimately absorb some degree of heat, we can make the assertion that the heat lost from the mixing of the hot and cold bodies of water, denoted $T_{Hot}$ and $T_{Cold}$, we can distinguish the degree of heat absorbed by the calorimeter as being the difference between the median value of the water temperature, $T_{Warm}$ and the final temperature $T_{Final}$, through: $$T_{Cal} = \frac{T_{Hot} - T_{Cold}}{2} - T_{Final}$$

This, as a result, will allow us to estimate the specific heat of the calorimeter through the equation $q_{p} = C_{p}\Delta T$ which we use to equate the water and the calorimeter through: $C_{p, water}\Delta T_{water} = C_{p, cal}T_{cal}$ where $\Delta T_water$ is equal to $T_{Hot} - T_{Final}$ and $T_{Cal}$ refers to the difference between the warm and final temperature mentioned above. Utilizing this equation, we can predict the specific heat of the calorimeter.

In the second part of the experiment, three reactions take place with the intent of gather the enthalpy values of mixing for cyclohexane-acetone, water-ethanol, and cyclohexane-ethanol reactions. These reactions were carried out through a process similar to that of the first reaction. In each sub-reaction, each of the two reactants were added in inversely proportional volumes, incrementing between 50 and 5mL. The reactant possessing the greatest volume for each respective trial was poured in first and stirred, with its temperature being taken immediately after. The second reactant was measured out and had its temperature taken before being added to the calorimeter, thus establishing two seperate temperatures to be measured. The two reactants were then mixed together, with their resulting final temperature being taken for each successive trial after roughly 20-30 seconds a piece. These varying temperatures aid in accomplishing the objective of the experiment as will be demonstrated in the latter portions of this report.

```python
import numpy as np
from matplotlib import pyplot as plt
## Part 1:
c_water = 4.184
mmw = 18.015
cpm_water = round(4.14 * mmw, 2)
mass_cal = 401.3
group1 = {
  'Th': [41.5, 36.4, 34.7],
  'Tc': [21.11, 23.3, 22.5],
  'Mh': [19.5, 18.6, 19.12],
  'Mc': [19.0, 19.6, 20.1],
  'Tf': [27.0, 29.1, 28.0]
}
group2 = {
  'Th': [50.2, 46.15, 41.85],
  'Tc': [21.15, 22.91, 23.23],
  'Mh': [27.42, 27.7, 28.62],
  'Mc': [29.96, 29.24, 28.52],
  'Tf': [27.76, 29.45, 28.44]
}
group3 = {
  'Th': [47.8, 38.8, 35.5],
  'Tc': [20.83, 22.6, 22.55],
  'Mh': [29.5, 29.4, 29.63],
  'Mc': [28.7, 30.9, 30.1],
  'Tf': [28.27, 27.68, 27.4]
}
## Functions
def dt(tf, tc):
    return tf - tc
def qhot(mhot, thot, tfinal):
    dt = thot - tfinal
    return mhot*dt*c_water
def qcold(mcold, tcold, tfinal):
    dt = tfinal - tcold
    return mcold*dt*c_water
def c_cal(qhot, qcold, tcal):
    qcal = qhot - qcold
    return qcal/tcal

## q_cal = (qhot - qcold)/(Tfinal - Tcold)

## Group 1
qh1 = [round(qhot(x,y,z), 2) for x,y,z in zip(group1['Mh'], group1['Th'], group1['Tf'])]
qc1 = [round(qcold(x,y,z), 2) for x,y,z in zip(group1['Mc'], group1['Tc'], group1['Tf'])]
dt1 = [round(dt(x,y), 2) for x,y in zip(group1['Tf'], group1['Tc'])]
cc1 = [round(c_cal(x,y,z), 2) for x,y,z in zip(qh1, qc1, dt1)]
print("Group 1: \n", qh1, "qhot in J\n", qc1, "qcold in J\n", dt1, 'In K\n', cc1, "In J/K")
############

## Group 2
qh2 = [round(qhot(x,y,z), 2) for x,y,z in zip(group2['Mh'], group2['Th'], group2['Tf'])]
qc2 = [round(qcold(x,y,z), 2) for x,y,z in zip(group2['Mc'], group2['Tc'], group2['Tf'])]
dt2 = [round(dt(x,y), 2) for x,y in zip(group2['Tf'], group2['Tc'])]
cc2 = [round(c_cal(x,y,z), 2) for x,y,z in zip(qh2, qc2, dt2)]
print("\nGroup 2: \n", qh2, "qhot in J\n", qc2, "qcold in J\n", dt2, 'In K\n', cc2, "In J/K")
############

## Group 3
qh3 = [round(qhot(x,y,z), 2) for x,y,z in zip(group3['Mh'], group3['Th'], group3['Tf'])]
qc3 = [round(qcold(x,y,z), 2) for x,y,z in zip(group3['Mc'], group3['Tc'], group3['Tf'])]
dt3 = [round(dt(x,y), 2) for x,y in zip(group3['Tf'], group3['Tc'])]
cc3 = [round(c_cal(x,y,z), 2) for x,y,z in zip(qh3, qc3, dt3)]
print("\nGroup 3: \n", qh3, "qhot in J\n", qc3, "qcold in J\n", dt3, 'In K\n', cc3, "In J/K")

tmax = np.array([41.5, 50.2, 47.8])
ccmax = np.array([max(cc1), max(cc2), max(cc3)])
```

```python
print("Graph of Heat Capacity vs. Tmax\n")

def bestfit(tmp, cc):
    m = (((mean(tmp)*mean(cc)) - mean(tmp*cc)) /
         ((mean(tmp)*mean(tmp)) - mean(tmp*tmp)))
    b = mean(cc) - m*mean(tmp)
    return m, b
m, b = bestfit(tmax, ccmax)


regression = [(m*x)+b for x in tmax]
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(tmax, ccmax, '*b', label='Experiemntal')
plt.plot(tmax, regression, 'red', label='Line of Best Fit')
plt.xlabel('Temperature of Hot Water')
plt.ylabel('Specific Heat in J/K')
plt.title('Heat Capacity vs. Tmax')
plt.legend()
for xy in zip(tmax, ccmax):
    ax.annotate('(%s, %s)' % xy, xy=xy, textcoords='data')
plt.show()


## Reference: https://pythonprogramming.net/how-to-program-best-fit-line-machine-learning-tutorial/
## https://stackoverflow.com/questions/22272081/label-python-data-points-on-plot
```

```python
### Part 2:
## qtotl = mh*cs(Tw-thot)
## qwarm = mc*cs*(twarm-tcold)
## Ccal = qc/(tw-tc)
## use densities to find the mc and the mh
r = 8.314
molwater, moleth, molcyc, molace = 18.02, 46.07, 84.16, 58.08
g1 = {
    ## Water
    'T_ini1' : [292.8, 293.6, 294.8, 294, 294.1, 294.1, 294.1, 294.9],
    'V_ini1' : [57, 48, 43, 38, 36, 30, 15, 5],
    ## Ethanol
    'T_ini2' : [293.5, 294, 294, 294, 295, 295, 294, 294],
    'V_ini2' : [3, 12, 17, 22, 24, 30, 45, 55],
    'T_final' : [294.5, 298.5, 300.5, 300.6, 300.6, 300.4, 300.6, 295.8]
}
g2 = {
    ## Cyclo
    'T_ini1' : [293.38, 292.32, 293.33, 294.33, 292.56, 293.2, 295.85],
    'V_ini1' : [55, 50, 45, 40, 35, 30, 10],
    ## Acetone
    'T_ini2' : [296.03, 294.4, 293.91, 294.45, 293.85, 293.97, 293.2],
    'V_ini2' : [5, 10, 15, 20, 25, 30, 50],
    'T_final' : [290.04, 286.96, 286.5, 286.21, 285.84, 285.77, 289.8]
}
g3 = {
    ## Cyclo
    'T_ini1' : [293.36, 293.46, 293.8, 294.35, 293.93],
    'V_ini1' : [50, 45, 25, 15, 5],
    ## Ethanol
    'T_ini2' : [294.02, 294.06, 293.75, 293.96, 294.15],
    'V_ini2' : [10, 15, 35, 45, 55],
    'T_final' : [291.42, 291.28, 290.12, 291.14, 293.54]
}
def molfr(mol1, mol2):
    return (mol1 / (mol1 + mol2))

## Group 1: 
dens_water1 = 0.998
dens_eth1 = 0.785
mass_water1 = [round(x * dens_water1, 2) for x in g1['V_ini1']]
mass_eth1 = [round(x * dens_eth1, 2) for x in g1['V_ini2']]
molw1 = [x / molwater for x in mass_water1]
moleth1 = [x / moleth for x in mass_eth1]
## Mole Fractions
mfw1 = [molfr(x,y) for x,y in zip(molw1, moleth1)]
mfeth1 = [molfr(x,y) for x,y in zip(moleth1, molw1)]



## Group 2:
dens_cyc2 = 0.783
dens_ace2 = 0.781
mass_cyc2 = [round(x * dens_cyc2, 2) for x in g2['V_ini1']]
mass_ace2 = [round(x * dens_ace2, 2) for x in g2['V_ini2']]
molcyc2 = [x / molcyc for x in mass_cyc2]
molace2 = [x / molace for x in mass_ace2]
mfcyc2 = [molfr(x,y) for x,y in zip(molcyc2, molace2)]
mface2 = [molfr(x,y) for x,y in zip(molace2, molcyc2)]


## Group 3:
dens_cyc3 = 0.783
dens_eth3 = 0.785
mass_cyc3 = [round(x * dens_cyc3, 2) for x in g3['V_ini1']]
mass_eth3 = [round(x * dens_eth3, 2) for x in g3['V_ini2']]
molcyc3 = [x / molcyc for x in mass_cyc3]
moleth3 = [x / moleth for x in mass_eth3]
mfcyc3 = [molfr(x,y) for x,y in zip(molcyc3, moleth3)]
mfeth3 = [molfr(x,y) for x,y in zip(moleth3, molcyc3)]

## Ideal Solvent Free Energy
## Delta G = R*T(x1ln(x1) + x2ln(x2))
def idealE(mf1, mf2, tf):
    return r * tf * ((mf1*np.log(mf1))+(mf2*np.log(mf2)))
idealg1 = [round(idealE(x,y,z), 2) for x,y,z in zip(mfw1, mfeth1, g1['T_final'])]
print("Group 1 Ideal Mixing Free Energy: ", idealg1)
idealg2 = [round(idealE(x,y,z),2) for x,y,z in zip(mfcyc2, mface2, g2['T_final'])]
print("Group 2 Ideal Mixing Free Energy: ", idealg2)
idealg3 = [round(idealE(x,y,z), 2) for x,y,z in zip(mfcyc3, mfeth3, g3['T_final'])]
print("Group 3 Ideal Mixing Free Energy: ", idealg3)


## Total Heat
cpm_ace, cpm_cyc, cpm_eth, cpm_w = 126.38, 154.37, 112.40, 72.23
def csmix(mf1, cp1, mf2, cp2):
    return ((mf1*cp1) + (mf2*cp2))
cs1 = [csmix(a,cpm_w,c,cpm_eth) for a, c in zip(mfw1, mfeth1)]
cs2 = [csmix(a,cpm_cyc, c, cpm_ace) for a, c in zip(mfcyc2, mface2)]
cs3 = [csmix(a,cpm_cyc, c, cpm_eth) for a, c in zip(mfcyc3, mfeth3)]

def qtot(mh, cs, tw, th):
    dt = tw - th
    return mh * cs * dt * -1

def qwarm(mc, cs, tw, tc):
    dt = tw - tc
    return mc * cs * dt
## g1 total heat:
twg1 = [(x+y)/2 for x,y in zip(g1['T_ini1'], g1['T_ini2'])]
qtotg1 = [round(qtot(a,b,c,d), 2) for a,b,c,d in zip(mass_eth1, cs1, twg1, g1['T_ini2'])]
qwarmg1 = [round(qwarm(a,b,c,d), 2) for a,b,c,d in zip(mass_water1, cs1, twg1, g1['T_ini1'])]
print("Group 1 Heat from Mix: ",qwarmg1)
## qwarm = -qtotal - qcal

## g2 total heat:
twg2 = [(x+y)/2 for x,y in zip(g2['T_ini1'], g2['T_ini2'])]
qtotg2 = [round(qtot(a,b,c,d),2) for a,b,c,d in zip(mass_ace2, cs2, twg2, g2['T_ini2'])]
qwarmg2 = [round(qwarm(a,b,c,d), 2) for a,b,c,d in zip(mass_cyc2, cs2, twg2, g2['T_ini1'])]
print("Group 2 Heat from Mix: ",qwarmg2)
## g3 total heat:
twg3 = [(x+y)/2 for x,y in zip(g3['T_ini1'], g3['T_ini2'])]
qtotg3 = [round(qtot(a,b,c,d),2) for a,b,c,d in zip(mass_eth3, cs3, twg3, g3['T_ini2'])]
qwarmg3 = [round(qwarm(a,b,c,d), 2) for a,b,c,d in zip(mass_cyc3, cs3, twg3, g3['T_ini1'])]
print("Group 3 Heat from Mix: ",qwarmg3)

## Net Delta H_mix:

dHg1 = [round(x-y, 2) for x,y in zip(qwarmg1, idealg1)]
dHg2 = [round(x-y, 2) for x,y in zip(qwarmg2, idealg2)]
dHg3 = [round(x-y, 2) for x,y in zip(qwarmg3, idealg3)]
print("Group 1 Delta H: ", dHg1, '\nGroup 2 Delta H: ', dHg2, '\nGroup 3 Delta H: ', dHg3)



## References: 
## http://www.ddbst.com/en/EED/PCP/DEN_C174.php (Water density @ 20C)
## http://www.ddbst.com/en/EED/PCP/DEN_C11.php (Ethanol density @ 26.85C)
## http://www.ddbst.com/en/EED/PCP/DEN_C50.php (Cyclohexane density @ 20.38C)
## http://www.ddbst.com/en/EED/PCP/DEN_C4.php (Acetone density @ 25C)

## http://www.ddbst.com/en/EED/PCP/HCP_C4.php (Acetone Sp. Heat @ 293.15K)
## http://www.ddbst.com/en/EED/PCP/HCP_C50.php (Cyclohexane Sp. Heat @ 295.43K)
## http://www.ddbst.com/en/EED/PCP/HCP_C11.php (Ethanol Sp. Heat @ 298.15K)
## http://www.ddbst.com/en/EED/PCP/HCP_C174.php (Water Sp. Heat @ 288.33K)
```

$\textbf{Discussion}$

Compared to the ideal values, we can see that the enthalpy of mixing has a large impact on the overall free energy of the reaction. Across all three groups, the ideal mixing free energy, which represents the expression $-T\Delta S$, from the original Gibbs Free Energy equation, possessed negative values. This implies spontaneity across each case in that $\Delta S_{mixing}$ can be expressed as the negative change in Gibbs Free Energy with respect to temperature, through the expression: $\Delta S_{mixing} = -\Big(\frac{\partial G_{mixing}}{\partial T}\Big)_{P}$, which is also equivalent to the expression $-nR \sum_{i} x_{i}\ln x_{i}$.$^{4}$ Since this expression indicates a negative change in $\Delta G$, we would expect positive entropy values to indicate spontaneity-- as was the case through the values calculated (considering the fact that the free energy values calculated are related to $T\Delta S_{mixing}$ through a factor of -1.) The inclusion of enthalpy values into consideration then becomes the dictator of the overall spontaneity of the reaction, as its addition to $-T\Delta S$ will ultimately decide whether or not $\Delta G$ remains negative in the end.

In the final plots constructed, comparing the mole fraction of ethanol to the entropy of the mixture behave as expected. The predicted model suggests that the relationship between the mole fraction and the entropy will be positive up until the mole fraction reaches roughly 0.5, and from there on it will become negative, as can be seen in these two graphs. This parabola shape occurs due to the fact that G experiences the greatest decrease when the mole fractions of the reactants are equal. From each of our experiments, we can observe that the while our plots may be similar to the ideal values, the ultimate ratio where G is decreased the most occurs at different mole fractions based upon the components. For instance, the last two plots show the graphs of the mole fractions of cyclohexane and ethanol with respect to the entropy. We can see that the peak on these two graphs occurs when cyclohexane's mole fraction is roughly around 0.65 and ethanol's is roughly around 0.35, rather than 0.5 which would be the case in an ideal solution. The ratios between mole fractions and peak entropy will differ based upon the reactants involved in the reaction.
The differences between the values in tables two and three, which are represented by groups 1 and 2 respectively in this experiment, show very similar relations between mole fractions and entropy values. As can be seen in the two plots for acetone and ethanol, the peak of the graph occurs almost exaclty at 0.5, indicating that these two reactions most accurately simulate ideal mixing.

$\textbf{Error Analysis}$

From analyzing the results collected during this experiment, it is clear that there were errors present within our values. One of the more apparent conflicts with our experiment is the fact that in part 2 of this experiment, the initial temperatures of some of the reactants were altering between trials as much as three to four degrees, which undoubtedly would have resulted in altered values. Additionally, it is worth mentioning that in the first trial for gorup 2, the cyclohexane and acetone volumes were mixed without the temperature of cyclohexane under stirring being taken, likely causing an altered result for the first trial. 
Other sources of error may include the possibility of the calorimeter retaining residual heat from previous trials as further ones were tested, altering the resulting temperatures from their actual values.
