$\textbf{Abstract:}$
The purpose of this experiment was to observe the behavior of air at varying temperatures and the subsequent effects on volume while maintaining constant pressure. The experiment was performed utilizing a water bath and several graduated cylinders containing a set amount of air within them and observing the changes in volume as the water temperature changed. 

$\textbf{Introduction:}$
The objective of this experiment was ultimately to determine the volume of the air within the bubble present within each graduated cylinder tested. This is possible due to Dalton's Law of Partial Pressure, which allows for us to equate the pressure exerted onto the bubble with the pressure within the bubble. Dalton's Law states that the total pressure must be equivalent to the summation of all individual partial pressures in some mixture. (1) Furthermore, his work tells us that for any expansion of fluid that occurs, the expansion of air is proportionally less. (2) This means that the total pressure outside of the bubble, which in our case is the pressure of the room plus the pressure of the fluid, must equal the pressure on the inside of the bubble, which is the pressure of the air plus the vapor pressure of water. The reason that the equivalent internal pressure of the bubble is not simply the pressure of the air can be explained through first defining vapor pressure. Vapor pressure is the temperature dependent equilibrium pressure of a liquid.(3) The vapor pressure of a water plays an important factor in our experiment, as the overall temperature of the water rises, the subsequent pressure formed from water vapor will also rise, which must be accounted for in each measurement involving a change in temperature.

The mathematical approach to obtaining the internal pressure, and subsequently the volume and moles of air, will be expanded on further in this report.

As far as the methodology of this experiment goes, the experiment was conducted by dropping three graduated cylinders with an initial pocket of air in them into a tank of water. The initial temperature, volume of air, and distance between the water level and the air bubble were all recorded. This step was followed up with alterations of the water temperature with subsequent measuring of the volume of air and distance between the water level and the bubble. After several different temperatures were taken, the results were analyzed, and the average volumes measured were used to calculate compression factors. The results showed mostly accurate values, with group 2 possessing the greatest deviations from the expected values.

## Q 3.2:
Dalton's Law of Partial Pressure states that: 

$$P_{total} = \sum P_a + P_b + \cdots +P_n$$

From this, we can find the pressure exerted on the air bubble within the water by equating it to the pressure from the outside

$$P_{total} = P_{air} + P_{water} = P_{room} + P_{fluid}$$

Fluid pressure is defined, as seen in reference 4, as:

$$P = \rho g h$$

This allows us to reformat our previous pressure equation into:

$$P_{total} = P_{room} + \rho_{water} g h$$

On the day of the experiment, the pressure of the room was recorded as 746mmHg. In order to find the most accurate pressure, it was also necessary to format our calculations using the coldest temperature, or 0$^{\circ}$C
The distance between the water level and the air bubble at this temperature was measured to be 162mm
Plugging all of this into our equation, we get:

$$P_{total} = 746mmHg + \frac{1000kg}{m^3} \times \frac{9.8m}{s^2} \times 0.162m$$


Reference:
4) https://www.grc.nasa.gov/WWW/K-12/WindTunnel/Activities/fluid_pressure.html

```python
pfluid = 1000*9.8*0.162
print("pfluid: ", pfluid)
## Because pfluid is currently in N/m^2, or Pascals, it must be converted into mmHg
## 101.325kPa = 760mmHg
pfluid /= 1000
pfluid = (pfluid/101.325) * 760
print("Pressure of fluid in mmHg: ", pfluid)
## From this, we find that pfluid is 11.9mmHg
## Now, plugging this into our Ptotal equation, we find that Ptotal is equal to 746 + 11.9, or 757.9mmHg
```

From the literature, reference 5, we find the vapor pressure of water at 0$^{\circ}$C to be 4.6mmHg
From Dalton's Law, we know that the pressure of the air bubble must be equal to $P_{total}$, or 757.9mmHg, therefore, to find the pressure of the air, we must subtract the 4.6mmHg from our $P_{total}$, as the equation is:

$$P_{air} = P_{total} - P_{water}$$

This gives us 753.3mmHg, which can be converted into atmospheres via $\frac{753.3mmHg}{760mmHg}$

Then with this value, we can calculate the number of moles of air via:

$$n_{air} = \frac{P_{air}V_{air}}{RT_{T = 0C}}$$

Reference:
5) http://intro.chem.okstate.edu/1515SP01/Database/VPWater.html

```python
pair = 753.3/760
## With this, we can calculate the number of moles of air through: PairVair/RT = nair
nair = pair*g1v1[2]/(R*tgroup1[2])
print("Number of moles of air: ", nair)
## From this, we find the number of moles of air present within the bubble to be 0.002078mol of air at a volume of 47mL
```

## Q 3.3:
With the new estimated moles of air calculated, we can calculate the volume of air at each temperature, correcting for contributions from water vapor and the actual pressure. To do this, we must first calculate the $P_{air}$ at each temperature first, utilizing the displaced distance in each trial, d.


```python
n = nair
vapor1 = {
    "23": 21.1,
    "15": 12.8,
    "0": 4.6,
    "29": 30,
    "43": 64.8}
vapor2 = {
    "22": 19.8,
    "12": 10.5,
    "0": 4.6,
    "30": 31.8,
    "14": 12}
vapor3 = {
    "20": 17.5,
    "8": 8,
    "0": 4.6,
    "35": 42.2,
    "55": 118
}
## Gives the Pair for a height, d given a specific vapor pressure in atmospheres
def pressure2(h, vp):
    pt = 1000*9.8*h/1000
    pt = (pt/101.325)*760
    pt += 746
    pt -= vp
    pt /= 760
    return pt
## Gets the volume of air in Liters
def volcalc(pair, t):
    v = R*t*n/pair
    v /= 1000
    return v
g1vols1, g1vols2, g1vols3 = [], [], []
g2vols1, g2vols2, g2vols3 = [], [], []
g3vols1, g3vols2, g3vols3 = [], [], []
g1p1, g1p2, g1p3 = [], [], []
g2p1, g2p2, g2p3 = [], [], []
g3p1, g3p2, g3p3 = [], [], []
## Calculated volumes
## Group 1 volumes:
for x, y, z in zip(tgroup1, vapor1.values(), g1dh1):
    p = pressure2(z, y)
    g1p1.append(p)
    v = volcalc(p, x)
    v *= 1000
    g1vols1.append(v)
for x, y, z in zip(tgroup1, vapor1.values(), g1dh2):
    p = pressure2(z, y)
    g1p2.append(p)
    v = volcalc(p, x)
    v *= 1000
    g1vols2.append(v)
for x, y, z in zip(tgroup1, vapor1.values(), g1dh3):
    p = pressure2(z, y)
    g1p3.append(p)
    v = volcalc(p, x)
    v *= 1000
    g1vols3.append(v)
print("Group 1 vol: ", "\n", g1vols1, "\n", g1vols2, "\n", g1vols3, "\n")

## Group 2 volumes:
for x, y, z in zip(tgroup2, vapor2.values(), g2dh1):
    p = pressure2(z, y)
    g2p1.append(p)
    v = volcalc(p, x)
    v *= 1000
    g2vols1.append(v)
for x, y, z in zip(tgroup2, vapor2.values(), g2dh2):
    p = pressure2(z, y)
    g2p2.append(p)
    v = volcalc(p, x)
    v *= 1000
    g2vols2.append(v)
for x, y, z in zip(tgroup2, vapor2.values(), g2dh3):
    p = pressure2(z, y)
    g2p3.append(p)
    v = volcalc(p, x)
    v *= 1000
    g2vols3.append(v)
print("Group 2 vol: ", "\n", g2vols1, "\n", g2vols2, "\n", g2vols3, "\n")

## Group 3 volumes:
for x, y, z in zip(tgroup3, vapor3.values(), g3dh1):
    p = pressure2(z, y)
    g3p1.append(p)
    v = volcalc(p, x)
    v *= 1000
    g3vols1.append(v)
for x, y, z in zip(tgroup3, vapor3.values(), g3dh2):
    p = pressure2(z, y)
    g3p2.append(p)
    v = volcalc(p, x)
    v *= 1000
    g3vols2.append(v)
for x, y, z in zip(tgroup3, vapor3.values(), g3dh3):
    p = pressure2(z, y)
    g3p3.append(p)
    v = volcalc(p, x)
    v *= 1000
    g3vols3.append(v)
print("Group 3 vol: ", "\n", g3vols1, "\n", g3vols2, "\n", g3vols3, "\n")


## Avg measured volumes for each group:
mg1v, mg2v, mg3v = [], [], []
for x, y, z in zip(g1v1, g1v2, g1v3):
    mg1v.append((x + y + z)/3)
for x, y, z in zip(g2v1, g2v2, g2v3):
    mg2v.append((x + y + z)/3)
for x, y, z in zip(g3v1, g3v2, g3v3):
    mg3v.append((x + y + z)/3)


## Avg  calculated volumes for each group:
fg1v, fg2v, fg3v = [], [], []
for x, y, z in zip(g1vols1, g1vols2, g1vols3):
    fg1v.append((x + y + z)/3)
for x, y, z in zip(g2vols1, g2vols2, g2vols3):
    fg2v.append((x + y + z)/3)
for x, y, z in zip(g3vols1, g3vols2, g3vols3):
    fg3v.append((x + y + z)/3)
print("Group 1 avg calculated vol: ", fg1v, "\n", "Group 2 avg calculated vol: ", fg2v, "\n", "Group 3 avg calculated vol: ", fg3v)
```

```python
## Q 3.4:
fg1p, fg2p, fg3p = [], [], []
## The pressure was already calculated when the volume was first being calculated
print("Group 1 pressure: ", "\n", g1p1, "\n", g1p2, "\n", g1p3, "\n")
print("Group 2 pressure: ", "\n", g2p1, "\n", g2p2, "\n", g2p3, "\n")
print("Group 3 pressure: ", "\n", g3p1, "\n", g3p2, "\n", g3p3, "\n")

## Avg pressures for each group:
for x, y, z in zip(g1p1, g1p2, g1p3):
    fg1p.append((x + y + z)/3)
for x, y, z in zip(g1p1, g1p2, g1p3):
    fg2p.append((x + y + z)/3)
for x, y, z in zip(g1p1, g1p2, g1p3):
    fg3p.append((x + y + z)/3)
print("Group 1 avg pressure: ", fg1p, "\n", "Group 2 avg pressure: ", fg2p, "\n", "Group 3 avg pressure: ", fg3p)
```

```python
## Q 3.5:
## Calculate the compression factor, z, at each temperature
## Utilizing the average pressures and volumes, we can find the average compression factor at each temperature
## The compression factors are found using the average calculated pressures, the measured volumes, and the temperatures 
## that relate the two
def zfactor(p, v, t):
    z = n*R*t/(p*v)
    return z
compg1, compg2, compg3 = [], [], []
zf1, zf2, zf3 = [], [], []
for x, y, z in zip(fg1p, mg1v, tgroup1):
    zf1.append(zfactor(x,y,z))
    temp = [z, zfactor(x, y, z)]
    compg1.append(temp)
for x, y, z in zip(fg2p, mg2v, tgroup2):
    zf2.append(zfactor(x,y,z))
    temp = [z, zfactor(x, y, z)]
    compg2.append(temp)
for x, y, z in zip(fg3p, mg3v, tgroup3):
    zf3.append(zfactor(x,y,z))
    temp = [z, zfactor(x, y, z)]
    compg3.append(temp)
print("\nTemp and Z factor for group 1: ", compg1, "\n", "\nTemp and Z factor for group 2: ", compg2, "\n", "\nTemp and Z factor for group 3: ", compg3)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```

Error Analysis:
It should be noted that as the measurements were recorded, there is a high probability for slight deviations in the measured heights and distances compared with the true values due to the inherent difficulty of obtaining these measurements. With that in mind, evaluation of the systematic and random errors can take place. In this lab, to measure the distances, a measuring stick was used to yield measurements in millimeters. This instrument possesses an uncertainty of 1mm, therefore, as mentioned in reference 6 on page 8, the uncertainty of this measurement is equal to the distance +/- 0.5mm. The volumes recorded in this lab were taken utilizing a graduated cylinder accurate to 100mL. As seen in reference 7, the uncertainty for a 100mL graduated cylinder is 1mL, meaning our volumes are accurate to +/- 1mL. 
As far as comparing our measured values to the calculated values goes, we can compare the calculated volumes against the measured volumes to gather a set of std deviations to provide insight into the measurement errors present within our results. 
The formula for calculating standard deviation is as follows:

\begin{equation}   
\sqrt{\frac{1}{N-1}\sum_{n=1}^{N} \left (x_i - \bar{x}\right)^2}
\end{equation}

Where $\bar{x}$ represents the average calculated volume of a given temperature and ${x}$ represents each measured volume.

References:

6) http://www.phy.olemiss.edu/~thomas/weblab/215_lab_items/35_215_APPENDIX__A.pdf

7) https://www.usna.edu/ChemDept/_files/documents/manual/apdxE.pdf

```python
from decimal import Decimal
def stdev(xbar, x):
    dev = lambda avg, val: (val - avg)**2 
    count = 0
    for j in x:
        count += (dev(xbar, j))
    std = ((1/2)*count)**.5
    return std
std1, std2, std3 = [], [], []
print(stdev(fg1v[0],(g1v1[0],g1v2[0],g1v3[0])), "\n")
## Calling the function
for x, y, z, avg in zip(g1v1, g1v2, g1v3, fg1v):
    std1.append(stdev(avg, [x, y, z]))
for x, y, z, avg in zip(g2v1, g2v2, g2v3, fg2v):
    std2.append(stdev(avg, [x, y, z]))
for x, y, z, avg in zip(g3v1, g3v2, g3v3, fg3v):
    std3.append(stdev(avg, [x, y, z]))
print("Group 1 volume std deviation: ", std1, "\nGroup 2 volume std deviation: ", std2, "\nGroup 3 volume std deviation: ", std3)
stdz1, stdz2, stdz3 = [], [], []
for x in zf1:
    stdz1.append(abs(1-x))
for x in zf2:
    stdz2.append(abs(1-x))
for x in zf3:
    stdz3.append(abs(1-x))
print("\nThe standard deviations between the experimental and theoretical compression factors: ")
print("Group 1: ", stdz1, "\nGroup 2: ", stdz2, "\nGroup 3: ", stdz3)

## Converted into percentages that is: 
dec1, dec2, dec3 = [str(round(x * 100, 2)) + '%' for x in stdz1], [str(round(x*100,2)) + '%' for x in stdz2], [str(round(x*100,2)) + '%' for x in stdz3]

print("\nCompression Factors converted to percentages: ", "\nGroup 1: ", dec1, "\nGroup 2: ", dec2, "\nGroup 3: ", dec3)

    

```
